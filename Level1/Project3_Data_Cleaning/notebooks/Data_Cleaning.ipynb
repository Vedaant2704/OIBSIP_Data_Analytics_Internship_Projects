```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
df = pd.read_csv("../data/raw_data.csv")
df.head()
```

```python
df.info()
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df.select_dtypes(include=np.number).iloc[:,0])
plt.title("Outlier Detection")
plt.show()
```

```python
df.info()
df.describe()
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)
```

## Data Cleaning Summary

- Missing values were identified and handled appropriately.
- Duplicate records were removed to maintain data integrity.
- Column names were standardized for consistency.
- Data types were corrected where required.
- Outliers were identified to prevent skewed analysis.

Clean data ensures accurate analysis and reliable results.
