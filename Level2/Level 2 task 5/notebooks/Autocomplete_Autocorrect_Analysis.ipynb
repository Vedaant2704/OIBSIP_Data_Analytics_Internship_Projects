```python
# Fix negative amounts automatically
df['Amount'] = df['Amount'].abs()
```

```python
# Cap extreme outliers automatically
upper_limit = df['Amount'].quantile(0.99)
df['Amount'] = df['Amount'].clip(upper=upper_limit)
```

```python
df['anomaly_flag'] = (df['Amount'] > df['Amount'].quantile(0.98)).astype(int)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop('Class', axis=1))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.boxplot(x=df['Amount'])
plt.title("Transaction Amount Validation & Correction")
plt.show()
```
